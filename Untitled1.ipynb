```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=["EIN", "NAME"], axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_counts = application_df.APPLICATION_TYPE.value_counts()
application_counts
```

```python
# Determine which values to replace if counts are less than ...?
replace_application = list(application_counts[application_counts < 500].index)

# Replace in dataframe
for app in replace_application:
    application_df.APPLICATION_TYPE = application_df.APPLICATION_TYPE.replace(app,"Other")
    
# Check to make sure binning was successful
application_df.APPLICATION_TYPE.value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df.CLASSIFICATION.value_counts()
classification_counts.head(20)
```

```python
# Determine which values to replace if counts are less than ..?
replace_class = list(classification_counts[classification_counts < 1000].index)

# Replace in dataframe
for cls in replace_class:
    application_df.CLASSIFICATION = application_df.CLASSIFICATION.replace(cls,"Other")
    
# Check to make sure binning was successful
application_df.CLASSIFICATION.value_counts()
application_df
```

```python
# Look at ASK_AMT value counts for binning
ask_amt_counts = application_df.ASK_AMT.value_counts()
ask_amt_counts.plot.density()
print(min(application_df.ASK_AMT),max(application_df.ASK_AMT),application_df.ASK_AMT.mean())
```

```python
# Establish the bins.
size_bins = [0,5000,100000,10000000000]
group_names = ["<5k","5k-100k","10M+"]

# size_bins = [0,5000,10000000000]
# group_names = ["<5000", "5000+"]

# Categorize spending based on the bins.
application_df["ASK_AMT_SIZE"] = pd.cut(application_df["ASK_AMT"], size_bins, labels=group_names)
application_df["ASK_AMT_SIZE"].value_counts()
```

```python
# Generate our categorical variable lists
application_df = application_df.drop(columns=["ASK_AMT"], axis=1)
application_df["ASK_AMT_SIZE"].value_counts()
```

```python
application_cat = application_df.dtypes[application_df.dtypes != "int64"].index.tolist()
application_cat
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)
```

```python
# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_df[application_cat]))
```

```python
# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(application_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
application_df = application_df.merge(encode_df, left_index=True, right_index=True).drop(columns=application_cat, axis=1)
application_df.head()
```

```python
# preprocess data intofeatures and target arrays
y = application_df.IS_SUCCESSFUL
X = application_df.drop("IS_SUCCESSFUL", axis=1)

#Split data  into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, stratify=y)
```

```python
# Scaling data

scaler = StandardScaler()

X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model
number_input_features = len(X_train_scaled[0])
nodes_hidden_layer1 = 48

nodes_hidden_layer2 = 6

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=nodes_hidden_layer1, activation="relu", input_dim=number_input_features))


# second hidden layer
nn.add(tf.keras.layers.Dense(units=nodes_hidden_layer2, activation="relu"))

# # Third hidden layer
# nn.add(tf.keras.layers.Dense(units=nodes_hidden_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
import os
from tensorflow.keras.callbacks import ModelCheckpoint

nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

os.makedirs("checkpoints_optimization_add_n2/", exist_ok=True)
checkpoint_path = "checkpoints_optimization_add_n2/weights.{epoch:02d}hdf5"
```

```python
cp_callback = ModelCheckpoint(filepath=checkpoint_path,verbose=0,save_weights_only=True,save_freq='epoch', period=5)
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[cp_callback])
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
